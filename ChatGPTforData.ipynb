```python
# =====================================
# TASK 3 - CHATGPT FOR DATA Q&A
# Simple and Easy Version
# =====================================

# Import Library
import pandas as pd

# Create Dataset
data = {
    "Name": ["Arun", "Priya", "Rahul", "Sneha"],
        "Math": [90, 85, 70, 95],
            "Science": [92, 80, 75, 98]
            }

df = pd.DataFrame(data)

# Display Dataset
print("Student Dataset:\n")
print(df)

# Save Dataset
df.to_csv("student_marks.csv", index=False)

# Question 1
top_math = df.loc[df["Math"].idxmax()]

print("\nTop Student in Math:")
print(top_math["Name"])

# Question 2
top_science = df.loc[df["Science"].idxmax()]

print("\nTop Student in Science:")
print(top_science["Name"])

# Question 3
average_math = df["Math"].mean()

print("\nAverage Math Marks:")
print(average_math)

# Convert Dataset to String
print("\nDataset as String:\n")
print(df.to_string())

# Create Q&A Report
qa_data = {
    "Question": [

              "Top Student in Math",
                    "Top Student in Science",
                            "Average Math Marks"
                                                ],

                                                    "Answer": [
                                                            top_math["Name"],
                                                                    top_science["Name"],
                                                                            average_math
                                                                                                ]
                                                                                                }

qa_df = pd.DataFrame(qa_data)

# Save Q&A Report
qa_df.to_csv("qa_report.csv", index=False)

print("\nQ&A Report Saved Successfully!")

# Conclusion
print("\nProject Completed Successfully!")
```
